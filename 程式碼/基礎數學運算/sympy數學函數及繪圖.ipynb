# sympy 數學函數簡介

>Sympy是用符號計算的Python套件。其目標是成為一個全功能的計算機代數系統，同時保持代碼簡潔、易於理解和擴展。它完全由Python寫成，不依賴外部程式庫。功能有支持符號計算、高精度計算、模式匹配、繪圖、解方程、微積分、組合數學、離散數學、幾何學、概率與統計、物理學等方面。而在Python眾多的科學運算套件裡，只有Sympy可以使用**代數**及其他各種**數學符號**來作運算的。因為Sympy開發目的就是成為一個全功能的代數系統，就比較合乎純數學（高中數學，大學數學）的計算環境。

Sympy數學符號庫（sym代表了symbol，符號的意思）包括了積分，微分方程，三角函數等各類數學運算方法，是工科最基本的數學函數庫，媲美matlab，而且要比math函數庫的計算精度來得精確。其內容包刮：

* 多項式的合併、展開、化簡。
* 微積分運算。
* 微分方程求解。
* 線性代數運算。
* 係數匹配。
* 繪圖。
* simplify簡化運算式，solve方程自動求解，limit求極限，diff求導，dsolve計算微分方程。
* intergrate積分計算：包含定積分、不定積分、雙重定積分和雙重不定積分。

進階方面，Sympy可以支援的內容包括但不限於：

1. 範疇論（Category Theory）。
2. 微分幾何（Differential Geometry）。
3. 常微分方程（ODE）。
4. 偏微分方程（PDE）。
5. 傅立葉轉換（Fourier Transform）。
6. 集合論（Set Theory）。
7. 邏輯計算（Logic Theory）。


```python
import sympy

#印出所有sympy函數名稱
print(dir(sympy))
```

```python
import math

#sympy的精度比math高
print('math,5開根號**2 = ', math.sqrt(5)**2)
print('sympy,5開根號**2 = ',sympy.sqrt(5)**2)

#常數pi, e
print('常數 pi=', sympy.pi)
print('sin(pi/6) = ', sympy.sin(sympy.pi/6))

#圓周率（pi）與尤拉常數(e)，在 SymPy 中以符號的方式進行運算，若是要算出實際的浮點數，可以使用 evalf() 這個函數
print('sym.pi的實際數字=sympy.pi.evalf()=', sympy.pi.evalf())
print('常數 E=sympy.E.evalf()=', sympy.E.evalf())

#三角函數
#角度 = 30度
r1 = math.radians(30)
d1 = math.degrees(sympy.pi/6)
print(r1,d1)
#d1 = sympy.degree((sympy.pi)/6)
print('math, sin(30度) = ', math.sin(sympy.pi/6))
print('sympy, sin(30度) = ', sympy.sin(sympy.pi/6))
print('sympy, sin(30度) = ', sympy.sin(r1))

#反三角函數： arc Sin(0.5) = pi/6
r2 = sympy.asin(0.5)
d2 = math.degrees(r2)
print('ashin(0.5)= radians =',r2)
print('ashin(0.5)= degrees =',d2)

#factorial(x)	傳回 x 階乘 =x!
d3 = sympy.factorial(5)
print('5!=', d3)

#絕對值
d5 = sympy.Abs(-7)
print('絕對值 = sympy.Abs(-7)=', d5)

# 自然指數 = sympy.exp(1)
print('自然指數 = sympy.exp(1) = ', sympy.exp(1))

#求20，16的最大公因數
print('求20，16的最大公因數=', sympy.gcd(20,16))

#四捨五入，無條件捨去，無條件進位ceiling
print('6.7的四捨五入 =', round(6.7))
print('6.7的無條件捨去 =', sympy.floor(6.7))
print('6.7的無條件進位 =', sympy.ceiling(6.7))

#5的3次方 sympy.Pow()
print('5的3次方=', 5**3)
print('5的3次方=', sympy.Pow(5,3))

#開根號
print('25開根號 = ', sympy.sqrt(25))
```

# 使用Sympy繪製$\sin{\theta}$和$\cos{\theta}$的圖型


```python
from sympy import *

#設定變數及符號並繪製sin圖型
x = symbols('x')
plot(sin(x))

#指定x範圍畫圖
plot(sin(x),(x,-5, 5))

#畫三條圖：y=x, y=sin(x), y=cos(x)
plot(x,sin(x),cos(x),(x,-5, 5))

#設定畫圖顏色
p1 = plot(x,sin(x),cos(x),(x,-5, 5),show=False)
p1[0].line_color = 'green'
p1[1].line_color = 'r'
p1.show()
```

# 使用Sympy繪製拋物線和方程式等圖型

繪製的圖型有：

1. 拋物線 $y=x^2$。
2. 以10為底指數函數 $y=10^x$。
3. 標準指數函數 $y=e^x$。
4. 橢圓函數 $x^2+y^2 = 4$。
5. 雙曲線函數: $\frac{x^2}{4-y^2} = 1$


```python
from sympy import *
x = symbols('x')

#拋物線：y=x**2
plot(x**2,(x,-2, 2))

#以10為底的指數函數：y=10**x
plot(10**x,(x,-2, 2))

#標準指數函數：y=e**x
plot(E**x,(x,-2, 2))

#三個圖畫在一起
p1 = plot(x**2,10**x,E**x,(x,-2, 2),show=False)
#設定顏色
p1[0].line_color = 'r'
p1[1].line_color = 'g'
#設定圖例legend
p1[0].label = '$x^2$'
p1[1].label = '$10^x$'
p1[2].label = '$e^x$'
#要打開 True
p1.legend = True

p1.show()
```

```python
#隱函數：implicit function
#可以把 "x**2+y**2 = 4" 轉換成 "y=sqrt(4-x**2)" 只剩單變數
#若方程式有雙變數，就要先設定雙變數：x, y = symbols('x y')
x, y = symbols('x y')
#建立：橢圓函數：x**2+y**2 = 4
plot_implicit(Eq(x**2 + y**2,4),(x,-3,3),(y,-3,3))

#建立：雙曲線函數:x**2/4-y**2 =1
plot_implicit(Eq(x**2/4-y**2, 1))
```

# 使用Sympy建立變數符號和函數表達式


```python
#建立符號變數x
x = symbols('x')
print('x=',x)

#建立符號變數x,y(其中 x >0)
x, y = symbols('x y')
x = symbols('x', positive = True)
print('x=',x)

#建立5個變數,x1,x2,x3,x4,x5
x = symbols('x1:6')
print('x1,....x5=',x)
```

# 展示Sympy如何處理多項式、方程式和函數，並顯示出標準的數學符號

有關[多項式](https://zh.wikipedia.org/wiki/%E5%A4%9A%E9%A0%85%E5%BC%8F)、[方程式](https://zh.wikipedia.org/wiki/%E6%96%B9%E7%A8%8B)和[函數](https://zh.wikipedia.org/wiki/%E5%87%BD%E6%95%B0)的區別及說明，可以參考[維基百科](https://zh.wikipedia.org/wiki/Wikipedia:%E9%A6%96%E9%A1%B5)。

至於Sympy是以下列的方式處理多項式、方程式和函數。

* 以plot指令處理多項式（polynomial）。
    > `poly(x**2+2*x+1)`  

* 以Eq指令處理方程式（Equation）。
    > `Eq(x**2+2*x+1, 0)`  

* 以Function指令處理函數（Function）。
    > `x = symbols('x')`  
    > `f = Function('f')(x)`  
    > `f = x**2+2*x+1`  
    > `#求值`  
    > `num = f.subs(x,2)`  

* 顯示出標準數學表達式。

    1. 先設定以latex格式來顯示數學式。
        > `printing.init_printing(use_latex=True)`

    2. 以IPython display類別裡的display指令在畫面顯示，print指令只能以文字方式顯示。
        > `f = poly(x**2+2*x+1)`  
        > `display(f)`


```python
#多項式：(x+2y+3)**2
#設定以latex格式把方程式輸出在畫面上
from sympy.interactive import printing
from IPython.display import display

printing.init_printing(use_latex=True)

x = symbols('x')

#一元多項式
f = poly(x**2+2*x+1)

display(f)

#一元多項式，相乘，展開
f = poly((x+3)*(x+5))
print(f)

#多項式相除
g = poly((x**3+x**2+x+1)/(x-2))
display(g)

#二元多項式
x,y = symbols('x y')
f = poly('(x+2*y+3)**2')
display(f)

#一元函數求對應值：f(x=2)
#求值f(x=3)，兩種寫法：(1) f.subs({x:3}。（2）f.subs(x,3)
f = Function('y')(x)
f = (x+2)**2
print('f(x=2)=',f.subs({x:3}))
print('f(x=2)=',f.subs(x, 3))

#二元函數求值：g(x=2,y=1)
g = Function('g')(x,y)
g = (x+2*y+3)**2
print('g(x=2,y=1) = ', g.subs({x:2, y:1}))
```
